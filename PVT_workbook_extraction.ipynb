# **Optimum Sep & Adjusted PVT Parameters**

```python
from ast import Index
# Initialize an empty dictionary to store the optimum separator conditions
optimum_separator = {
    "test" :    [],
    "pressure":     [],            # psia
    "temperature":     [],         # °F
    "total GOR":      [],          # scf/STB
    "stock tank API":     [],      # °API
    "formation volume factor": []
    }

# Find the index of the minimum 'formation volume factor' in the 'seperator_test' dictionary
minBo = seperator_test['formation volume factor'].index(min(seperator_test['formation volume factor']))

# Populate the 'optimum_separator' dictionary with values corresponding to the minimum 'formation volume factor'
optimum_separator['test'].append(seperator_test['test'][minBo])
optimum_separator['pressure'].append(seperator_test['pressure'][minBo])
optimum_separator['temperature'].append(seperator_test['temperature'][minBo])
optimum_separator['total GOR'].append(seperator_test['total GOR'][minBo])
optimum_separator['stock tank API'].append(seperator_test['stock tank API'][minBo])
optimum_separator['formation volume factor'].append(seperator_test['formation volume factor'][minBo])


optimum_separator
```

## **Correction Factors for Bo and Rs**

```python
# Calculate the correction factor for Oil Formation Volume Factor (Bo)
# using the optimum separator's FVF and the DLE's Bo at bubble point
Bo_correction_factor = optimum_separator['formation volume factor'][0]/DLE_Bo['Bo'][0]

# Calculate the correction factor for Solution Gas-Oil Ratio (Rs)
# using the optimum separator's total GOR and the DLE's Rs at bubble point
Rs_correction_factor = optimum_separator['total GOR'][0]/DLE_Bo['Rs'][0]

# Print the calculated correction factors
print('Bo correction factor = ', Bo_correction_factor)
print('Rs correction factor = ', Rs_correction_factor)
```

# **Adjusted PVT Table**

```python
# Initialize a dictionary to store the adjusted Oil Formation Volume Factor (Bo) and Solution Gas-Oil Ratio (Rs)
adjusted_Bo_Rs = {
        "pressure":     [],            # psia
        "Bo_DLE":       [],            # Scf/Stb
        "Bo_adjusted":  [],            # Bbl/Stb
        "Rs_DLE":       [],            # Scf/Stb
        "Rs_adjusted":  [],            # Scf/Stb
}

# Iterate through the DLE pressure data to calculate adjusted Bo and Rs values
for i in range(len(DLE['pressure'])):
  pressure_value = DLE['pressure'][i]
  # Append original DLE values for pressure, Bo, and Rs
  adjusted_Bo_Rs['pressure'].append(pressure_value)
  adjusted_Bo_Rs['Bo_DLE'].append(DLE['Bo'][i])
  adjusted_Bo_Rs['Rs_DLE'].append(DLE['Rs'][i])

  # Apply adjustments based on pressure relative to the bubble point pressure
  if pressure_value > DLE_Bo['pressure'][0]:
    # Above bubble point pressure: Adjust Bo using CCE relative volume
    adjusted_Bo_Rs['Bo_adjusted'].append(DLE['Bo'][i] * CCE['relative volume'][i])
    # Rs is constant above bubble point pressure (DLE_Bo['pressure'][0]), use DLE_Bo's Rs
    adjusted_Bo_Rs['Rs_adjusted'].append(DLE_Bo['Rs'][0])
  else:
    # Below or at bubble point pressure: Adjust Bo and Rs using correction factors
    adjusted_Bo_Rs['Bo_adjusted'].append(DLE['Bo'][i] * Bo_correction_factor)
    adjusted_Bo_Rs['Rs_adjusted'].append(DLE['Rs'][i] * Rs_correction_factor)

# Display the resulting dictionary with adjusted values
# adjusted_Bo_Rs
```
